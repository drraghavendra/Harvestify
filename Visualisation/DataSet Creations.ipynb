```python
def giveHeatMap(l,name):
    plt.clf()
    plt.figure(dpi = 200)
    plt.title("Farm Soil Moisture Level")
    figure = sns.heatmap(l,annot = True, xticklabels = False,yticklabels = False)
    plt.savefig(name+'.png')
    plt.close()
```
